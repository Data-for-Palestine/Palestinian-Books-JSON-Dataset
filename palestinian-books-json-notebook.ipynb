- Path

```python
json_file_path = '../input/palestinian-books-json-dataset/palestinian_books.json'
```

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

- Read the JSON file into a DataFrame

```python
books_df = pd.read_json(json_file_path)
```

- Display the DataFrame

```python
books_df
```

```python
books_data = books_df['books']
```

```python
books_df = pd.json_normalize(books_data, sep='_')
```

```python
import seaborn as sns
```

- Histograms for Numerical Columns

```python
numerical_columns = books_df.select_dtypes(include='number').columns
for column in numerical_columns:
    books_df[column].hist(bins=20, edgecolor='black')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram for {column}')
    plt.show()
```

- Box Plots for Numerical Columns

```python
for column in numerical_columns:
    sns.boxplot(x=books_df[column])
    plt.title(f'Box Plot for {column}')
    plt.show()
```
